**Collections counter**

Um Counter é uma subclasse de dict subclass para contagem de objetos hasheáveis. É uma coleção na qual elementos são armazenados como chaves de dicionário e suas contagens são armazenadas como valores de dicionário. Contagens podem ser qualquer valor inteiro incluindo zero e contagens negativas. A classe Counter é similar a sacos ou multiconjuntos em outras linguagens.

Os elementos são contados a partir de um iterável iterable ou inicializado a partir de um outro mapeamento mapping (ou contador).

**Collections — Tipos de dados de contêineres — documentação Python 3.9.12. Disponível em: <https://docs.python.org/pt-br/3.9/library/collections.html#:~:text=Um%20Counter%20%C3%A9%20uma%20subclasse%20de%20dict%20subclass>.**

# **Importando Dados**


```python
#importação dos dados
dados_cenipa = pd.read_csv("/content/Dataset/ocorrencias_cenipa.csv",sep =';',encoding='ISO-8859-1')
```

## **Análise 1 - Nível de ocorrência (Incidente, Acidente e Incidente Grave)**


```python
# Tipos de Ocorrência
dados_cenipa.head()
tipos = ['ACIDENTE', 'INCIDENTE','INCIDENTE GRAVE']

# Criando Gráfico 1 - Tipos de ocorrência
fig, ax = plt.subplots(figsize=(14, 6))
ax.set_ylabel('Número de Ocorrências\n')
ax.set_title('\nTipos de Ocorrências\n')
dados_cenipa.groupby(['tipo']).tipo.count().plot(kind='bar')
```

De acordo com a análise gráfica, podemos observar que a quantidade de casos que se qualificam como **"Incidentes"** **se destaca** entre os outros dois cenários, sendo responsável por **2.887 casos** dentro do período estudado, seguido respectivamente, por **"Acidente"** e **"Indicidente Grave"**, com **1.694** e **661** casos.

## **Análise 2 - Nível de ocorrência (Incidente, Acidente e Incidente Grave) por percentual**

```python
# Criando gráfico 2 - Tipos de ocorrência por percentual
dados_cenipa['tipo'].value_counts().plot(kind='pie',autopct='%0.2f%%' ,label="Tipo Ocorrência (%)\n" ,figsize=(5, 5))
```

É possível observar que o número de **incidentes**, que possui um grau **menor de gravidade**, é responsável por grande parte dos casos mapeados, representando **55.07%** do total.

Já os **acidentes** e os **incidentes graves**, dividem pouco **menos da metade** do restante, cerca de 45%.

De forma isolada, o tipo de ocorrência **menos frequente** são os **incidentes graves**, ocupando 12.61% do estudo geral.

## **Análise 5 - Ocorridos por ano**

```python
# Gráfico 5 - Extraindo ano do dataset
dados_cenipa['data'] = pd.to_datetime(dados_cenipa['data'])
dados_cenipa['Ano'] = dados_cenipa['data'].map(lambda x: x.year)

ocorrencias_por_ano = Counter(dados_cenipa['Ano'])
ano = (list(ocorrencias_por_ano.keys()))
ocorrencia_ano = (list(ocorrencias_por_ano.values()))

plt.style.use("ggplot")
fig,ax = plt.subplots(figsize=(20, 6))

x=dados_cenipa.Ano

plt.hist(x, density=False, bins=40)
mn, mx = plt.xlim()
kde = st.gaussian_kde(x)
plt.xlabel('Ano')
plt.title('Quantidade de Ocorrências por Ano')

plt.show()
```

É notório que o período possue uma certa constância de casos, não apresentando nenhum ano que fique visivelmente disparado.

De toda forma, ainda é aceitável dizer que no período de 2011 até 2013 foram mapeados maior número de acidentes:


*   **2011:** 576 casos (10,99%)
*   **2012:** 647 casos (12,34%)
*   **2013:** 653 casos (12,46%)

O que pode se remeter a segurança, já que a cada dia os mecanismos são melhorados.
